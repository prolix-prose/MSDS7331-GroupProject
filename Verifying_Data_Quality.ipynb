```python
import numpy as np
# replace '?' with -1, we will deal with missing values later
df_replace = df.replace(to_replace='?',value=-1)
```

```python
df_replace.info()
```

```python
df_replace.describe()
```

```python

# how many value have the -1 (which we set as the missing values) 
import numpy as np

# let's set those values to NaN, so that Pandas understand they are missing
df_replace = df_replace.replace(to_replace=-1,value=np.nan) # replace -1 with NaN (not a number)
print (df_replace.info())
df_replace.describe() # scroll over to see the values
```

```python
df_replace.median()
```

```python
# the 'fillna' function will take the given series (the output above)
# and fill in the missing values for the columns it has
df_imputed = df_replace.fillna(df_replace.median()) # note that to do this all values must be numeric
df_imputed.info()
```
